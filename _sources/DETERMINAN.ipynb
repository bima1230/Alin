#Matriks A (2x2)

```python
# importing Numpy package
import numpy as np

# creating a 2X2 Numpy matrix
n_array = np.array([[1, 0], [0, 1]])

# Displaying the Matrix
print("Numpy Matrix is:")
print(n_array)

# calculating the determinant of matrix
det = np.linalg.det(n_array)

print("\nDeterminant of given 2X2 matrix:")
print(int(det))
```

# Matriks E (kelipatan dari baris ke-2 D)


### Sifat 3 : Jika suatu baris ataupun kolom merupakan kelipatan dari baris yang lain maka determinan matrix tersebut adalah 0 (baris 2 adalah kelipatan dari baris 1)


```python
# importing Numpy package
import numpy as np

# creating a 3X3 Numpy matrix
n_array = np.array([[1, 2, 1],
					[2, 4, 2],
					[2, 1, 2]])

# Displaying the Matrix
print("Numpy Matrix is:")
print(n_array)

# calculating the determinant of matrix
det = np.linalg.det(n_array)

print("\nJika suatu baris adalah kelipatan dari baris yang lain maka determinan matrix tersebut adalah 0 \nbaris 2 adalah kelipatan dari baris 1 matriks D ")
print(int(det))
```

# Matriks G
### sifat 4 : Jika matrix tersebut adalah kelipatan sebanyak n * matrix maka untuk mencari determinan adalah (n^(sebanyak ordo)determinan matrix)contoh matrix g adalah kelipatan 2 kali dari matrix d : (2^3(det d))

```python
# importing Numpy package
import numpy as np

# creating a 3X3 Numpy matrix
n_array = np.array([[6, 2, 4],
					[2, 4, 2],
					[4, 2, 4]])

# Displaying the Matrix
print("Numpy Matrix is:")
print(n_array)

# calculating the determinant of matrix
det = np.linalg.det(n_array)

print("\nMatriks G merupakan kelipatan 2 dari matriks D maka kesimpulannya (2^3(det d))")
print(int(det))
```

### Sifat 5 : Jika terdapat matrix baru dengan nilai salah satu baris atau kolom adalah kelipatan 2 dari matrix sebelumnya maka dapat disimpulkan dengan 2*(determinan matrix)
contoh matrix i baris pertama adalah kelipatan 2 dari matrix a

contoh matrix j kolom pertama adalah kelipatan 2 dari matrix a

```python
i = np.array([[6, 8],
					[2, 1]])

print("Matriks i :")
print(i)
print("sifat 5 : jika terdapat matrix baru dengan nilai salah satu baris adalah kelipatan 2 dari matrix sebelumnya maka dapat disimpulkan dengan 2x(det a)")
det = np.linalg.det(i)
print("Determinan : ")
print(det.round())
```

```python
j = np.array([[6, 4],
					[4, 1]])
print("Matriks j :")
print(j)
print("sifat 6 : jika terdapat matrix baru dengan nilai salah satu kolom adalah kelipatan 2 dari matrix sebelumnya maka dapat disimpulkan dengan 2x(det a)")
det = np.linalg.det(j)
print("Determian : ")
print(det.round())
```

### Sifat 8 : untuk mencari determinan dari Hasil perkalian matrix O dengan matrix p yaitu dengan mengalikan determinan matrix O dan p contoh (det O = 1 , det p = 1 maka det q adalah 0)

```python
O = np.array([[2, 1],
					[1, 1]])

print(O)
det = np.linalg.det(O)
print(int(det))
```

```python
p = np.array([[1, 1],
					[1, 1]])

print(p)
det = np.linalg.det(p)
print(int(det))
```

```python
q=O@p
print(q)
det = np.linalg.det(q)
print(int(det))
```
